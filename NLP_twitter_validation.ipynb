```python
#drop id and location
df.drop(['id','location'],axis = 1 , inplace = True)
df
```

```python
df = df.loc[df['target'] != 'Irrelevant']
df
```

```python
df.reset_index(drop = True,inplace =True)
df
```

```python
#convert to numberic value
#positve 1
#negative -1
#neutral 0

df['target']= df['target'].map({'Positive':1,'Negative':-1,'Neutral':0})
df
```

```python
tweets = df.text
tweets
```

```python
tweets  = tweets.str.replace('[^a-zA-Z0-9]+',' ')
tweets
```

```python
#tokenize and stemming
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer  = SnowballStemmer('english')
tk = TweetTokenizer()
```

```python
tweets = tweets.apply(lambda X : [stemmer.stem(token.lower()) for token in tk.tokenize(X)]).apply(lambda token:' '.join(token))
tweets
```

```python
#remove stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw =stopwords.words('english')
sw
```

```python
tweets  = tweets.apply(lambda X:[token for token in tk.tokenize(X) if token not in sw ]).apply(lambda token:' '.join(token))
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(tweets)
print(X)
```

```python
vectorizer.get_feature_names_out()
```

```python
X.shape # 3931 unique words
        # 828 X values
```

```python
y = df['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 1)
```

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

rf = RandomForestClassifier(random_state = 1)
ad = AdaBoostClassifier()
sv = SVC()
knn = KNeighborsClassifier()

lst  = [rf,ad,sv,knn]
for i in lst :
  print('***********************************',i,'*****************************')
  i.fit(X_train,y_train)
  y_pred = i.predict(X_test)
  print(classification_report(y_test,y_pred))
```
